```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1_train = pd.read_csv(r'C:\Users\LENOVO\Downloads\train_wn75k28.csv')
```

```python
## Removing columns
d1_train.drop(['created_at', 'signup_date'], axis=1, inplace = True)
```

## Feature Engineering

```python
## Checking Null Value
d1_train.isnull().sum()
```

```python
## Imputing Null Value by Median

d1_train.loc[d1_train['products_purchased'].isnull()==True,'products_purchased'] = np.median(d1_train.products_purchased.dropna(axis=0))
```

```python
d1_train.isnull().sum()
```

```python
## Checking duplicate Observations-:
d1_train.duplicated().sum()
```

## Creation of X & y

```python
y_train = d1_train.buy
```

```python
d1_train
```

```python
X_train = d1_train.drop(['buy'], axis =1)
```

```python
d1_test = pd.read_csv(r'C:\Users\LENOVO\Downloads\test_Wf7sxXF.csv')
```

```python
d1_test.drop(['created_at', 'signup_date'], axis=1, inplace = True)
```

```python
X_test = d1_test
```

```python
X_test
```

```python
## SCaling & Transformation

from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()
```

```python
d1_train[['campaign_var_1', 'campaign_var_2', 'products_purchased']] = scale.fit_transform(d1_train[['campaign_var_1', 'campaign_var_2', 'products_purchased']])
```

```python
X_train
```

```python
## Balancing the data

from imblearn.over_sampling import SMOTE

smote = SMOTE()
```

```python
X_smote, y_smote = smote.fit_resample(X_train, y_train)
```

```python
from collections import Counter
print('Before Classes', Counter(y_train))
print('Actual Classes', Counter(y_smote))
```
